## Car Detection from a picture

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

Reading a image

```python
car_pic = cv2.imread(r'C:\Users\Binary\car_pic.jpg',1)
```

We need a xml file for CascadeClassifier. The xml file is learned and created for different applications. There are many videos showing how to create xml files needed for Cascade classifiers

```python
cars_cascade = cv2.CascadeClassifier(r'C:\Users\Binary\haarcascade_car.xml')
```

1. first a frame is read.
2. Cars are detected by using detectMultiScale()
3. Rectangle is formed for the car co-ordinates
4. Text is written over the rectangles


```python
cars = cars_cascade.detectMultiScale(car_pic, 1.15, 1)
for (x, y, w, h) in cars:
    cv2.rectangle(car_pic, (x, y), (x+w,y+h), color=(0, 255, 0), thickness=2)
    font = cv2.FONT_HERSHEY_DUPLEX
    cv2.putText(car_pic, 'Car', (x + 6, y - 6), font, 0.5, (0, 0, 255), 1)   
```

```python
plt.figure(figsize=(10,20))
plt.imshow(car_pic)
plt.savefig('detection.png')
```
